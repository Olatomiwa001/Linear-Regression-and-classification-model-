```python
import pandas as pd 
import numpy as np 
```

```python
df = pd.read_csv('very_dirty_data.csv')
df
```

```python
df.info()
```

```python
df.head(10)
```

```python
df.isnull().sum()
```

```python
df['Age'].fillna(df['Age'].median(), inplace= True)
df['Salary'].fillna(df['Salary'].mean(), inplace=True)
df.dropna(subset=['Email'], inplace=True)
```

```python
df
```

```python
df.isnull().sum()
```

```python
df['Gender'] = df['Gender'].replace({'M': 'Male', 'F': 'Female', 'UNKNOWN': np.nan})
df['Membership_Level'].replace({'Unknown': np.nan}, inplace=True)
```

```python
df.isnull().sum()
```

```python
df
```

```python
df.head(30)
```

```python
print("Duplicate Rows:", df.duplicated().sum())
```

```python
df.drop_duplicates(inplace=True)
```

```python
df
```

```python
print("Duplicate Rows:", df.duplicated().sum())
```

```python
df = df[df['Salary']> 0]
```

```python
df['Salary'] = np.where(df['Salary']> 1000000, 1000000, df['Salary'])
```

```python
df.head(40)
```

```python
df.isnull().sum()
```

```python
df['Signup_Date'] = pd.to_datetime(df['Signup_Date'], errors='coerce')
```

```python
df.head(40)
```

```python
df.isnull().sum()
```

```python
df['Name'] = df['Name'].str.strip()
df['Feedback'] = df['Feedback'].str.strip()
```

```python
df = df[df['Purchase_History']>= 0]
```

```python
df.isnull().sum()
```

```python
df.head(40)
```

```python
df = df.dropna()
```

```python
df.head(40)
```

```python
df.isnull().sum()
```

```python
df.to_csv ("Cleaned_data.csv")
print("Data cleaning complete. Cleaned file saved as cleaned_data.csv.")
```
